## 7. Implementing the Solution

### Multinomial Naive Bayes Classification

#### With a 20% test size

```
# splitting data into features and target variables

X = spambase.iloc[:,:-1].values
y = spambase['class'].values

# splitting the dataset into training and testing data

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)

# using the multinomial bayes classifier and making predictions

clf = MultinomialNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
# evaluating the model
print(confusion_matrix(y_test, y_pred))
print('Mean Accuracy:',round(np.mean(y_pred==y_test)*100,2),'%')
```

With a 20% test size the model has a mean prediction accuracy of 81.22%.

#### With a 30% test size

```
# splitting data into features and target variables

X = spambase.iloc[:,:-1].values
y = spambase['class'].values

# splitting the dataset into training and testing data

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)

# using the multinomial bayes classifier and making predictions

clf = MultinomialNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
# evaluating the model
print(confusion_matrix(y_test, y_pred))
print('Mean Accuracy:',round(np.mean(y_pred==y_test)*100,2),'%')
```

With 30% test size the model's accuracy has reduced slightly to 80.96%

### Improving the models perfomance through feature selection

```
# Feature selection can be done by dropping highly correlated features
# creating a copy for feature selection
spambase_df = spambase.copy()

# Create correlation matrix
corr_matrix = spambase_df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
```

```
spambase_df = spambase_df.drop(spambase_df[to_drop], axis=1)
```

```
# checking to see if any of the columns were dropped
spambase_df.shape
```

One column that was highly correlated has been dropped. We will now fit the data to MultinomialNB and see if the model's accuracy improves

```
# splitting data into features and target variables

X = spambase_df.iloc[:,:-1].values
y = spambase_df['class'].values

# splitting the dataset into training and testing data

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)

# using the multinomial bayes classifier and making predictions

clf = MultinomialNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
# evaluating the model
print(confusion_matrix(y_test, y_pred))
print('Mean Accuracy:',round(np.mean(y_pred==y_test)*100,2),'%')
```

Fitting the new data after feature selection the mean accuracy model prediction is at 81.11%, this accuracy is slightly lower than what we had earlier when we used the same test size sample. We can conclude that the first model without feature selection is better as it perfomed better.

### Improving the models perfomance through normalization

```
# making a copy of df for normalization
spambase_norm = spambase_df.copy()
spambase_norm
```

```
# splitting data into features and target variables
X = spambase_norm.iloc[:,:-1].values
y = spambase_norm['class'].values

# Create normalizer
normalizer = Normalizer(norm='l2')

# Transform feature matrix
normalizer.transform(X)

# splitting the dataset into training and testing data

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)

# using the multinomial bayes classifier and making predictions

classifier = MultinomialNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
# evaluating the model
print(confusion_matrix(y_test, y_pred))
print('Mean Accuracy:',round(np.mean(y_pred==y_test)*100,2),'%')
```

Even after normalization the mean accuracy of our model still remains at 81.11%
